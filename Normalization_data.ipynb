```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True, max_features = 300)
```

```python
summary=imdb.SUMMARY.values
```

```python
print((summary[0]))
```

```python
print((summary[0][1]))
```

```python
X=cv.fit_transform(summary)
```

```python
print(type(X))
```

```python
new_sum=pandas.DataFrame(X.todense())
```

```python
new_sum.head(10)
```

```python
imdb=imdb.drop(['SUMMARY','TITLE'],axis=1)
```

```python
imdb
```

```python
transform_result = pandas.concat([imdb, new_sum], axis=1)
```

```python
transform_result.head(10)
```

```python
transform_result.to_csv("Naive_imdb.csv", sep='\t', encoding='utf-8')
```
